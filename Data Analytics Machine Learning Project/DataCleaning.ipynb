**Create Pandas DF's for each of the Stocks, and print the first few entries:**

```python
import pandas as pd

# Read the Stock Data CSVs file into a pandas DataFrame
AAPL_df = pd.read_csv("AAPL_stock_data.csv")
SP500_df = pd.read_csv("SP500_stock_data.csv")
PFE_df = pd.read_csv("PFE_stock_data.csv")

AAPL_df.head()
```

**Remove any rows with empty / non existent values:**

```python
AAPL_df = AAPL_df.dropna()
SP500_df = SP500_df.dropna()
PFE_df = PFE_df.dropna()
```

**Plot the closing prices of each stock each day, as well as the moving averages:**

```python
import matplotlib.pyplot as plt

for df in (AAPL_df, SP500_df, PFE_df):

    # Plotting
    plt.figure(figsize=(10, 6))  # Adjust figure size as needed

    # Plot closing price
    plt.plot(df['date'], df['close'], color='black', linewidth=2.5, label='Closing Price')

    # Plot moving averages
    plt.plot(df['date'], df['50_ma'], color='red', linewidth=1.0, label='50-day MA')
    plt.plot(df['date'], df['100_ma'], color='green', linewidth=1.0, label='100-day MA')
    plt.plot(df['date'], df['200_ma'], color='blue', linewidth=1.0, label='200-day MA')

    # Customize plot
    plt.title('Closing Price and Moving Averages')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)

    # Show plot
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Remove dates from df for correlation matrix

for df in (AAPL_df, SP500_df, PFE_df):
    # Remove dates from df for correlation matrix
    df = df.drop(columns=['date'])
    # Calculate correlation matrix
    corr_matrix = df.corr()

    # Plotting the correlation matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.4f')
    plt.title('Correlation Matrix')
    plt.show()
```
